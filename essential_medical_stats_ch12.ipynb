## Checking for normality

Idea: proportion of data falling below certain quantiles should match that predicted by standard normal distribution

```python
from scipy.stats import norm
```

```python
# order the data and calculate quantiles
mean_est = x.mean()
var_est = x.var()
```

```python
quantiles = np.array([n / x.shape[0] for n in range(0, x.shape[0])])
```

```python
quantiles.shape, x.shape
```

```python
probits = norm.ppf(quantiles)
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(x, probits)
plt.show()
```

## Uniformly distributed variables

```python
x = np.random.rand(2**16)
xvals = np.arange(0, 1, 1/1000)
yvals2 = 0.5 / np.sqrt(xvals)
```

```python
plt.figure(figsize=(12, 8))
plt.hist(x, density=True)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.hist(x**2, density=True, bins=64)
plt.plot(xvals[1:], yvals2[1:])
plt.show()
```
